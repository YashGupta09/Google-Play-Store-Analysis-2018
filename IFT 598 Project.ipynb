```python
import pandas as pd
import numpy as np
```

```python
appDF = pd.read_csv('Data/googleplaystore.csv', delimiter=',', header=0)
appDF.head(10)
```

```python
appDF.shape
```

```python
#Removing 'App', 'Price', 'Genres', 'Last_Updated', 'Current_Ver' & 'Android_Ver' cols
appDF2 = appDF.drop(['App', 'Price', 'Genres', 'Last_Updated', 'Current_Ver' ,'Android_Ver'], axis=1)
appDF2.head(10)
```

```python
appDF2.shape
```

```python
#Dropping NaN values
appDF3 = appDF2.dropna(how='any')
```

```python
pd.options.mode.chained_assignment = None
```

```python
#Removing k and M from Size column
def cleanSize(size):
    letter = size[-1:]
    if(letter == 'k'):
        return (float(size[:-1]) * 1000)
    elif(letter == 'M'):
        return (float(size[:-1]) * 1000000)
    else:
        return "VWD"
        
appDF4 = appDF3.copy()
appDF4['Size'] = appDF4['Size'].map(cleanSize)
```

```python
appDF4.head(10)
```

```python
appDF4.Size.unique()
```

```python
appDF4.Size.describe()
```

```python
appDF4.dtypes
```

```python
appDF4.Reviews = appDF4.Reviews.astype(int)
appDF4.Category = appDF4.Category.astype('category')
appDF4.dtypes
```

```python
#Cleaning Installs column
appDF5 = appDF4.copy()
appDF5.Installs = appDF5.Installs.map(lambda x: x.replace(',', '').replace('+', '')).astype(int)

#Encoding Installs, technique - encoding labels
installsList = sorted(appDF5.Installs.unique())
installsDict = {'Installs': {}}
for i in range(len(installsList)):
    installsDict['Installs'][installsList[i]] = i
appDF5.Installs = appDF5.Installs.map(lambda x: installsDict['Installs'].get(x, 100))
appDF5.head(20)
```

```python
installsDict
```

```python
appDF5.Installs.value_counts()
```

```python
#One-hot encoding of Type column
appDF6 = appDF5.copy()
appDF6.Type = appDF6.Type.astype('category')
appDF6 = pd.get_dummies(appDF6, columns=['Type'], prefix=['Type'])
appDF6.head()
```

```python
appDF6.dtypes
```

```python
appDF6.Content_Rating.value_counts()
```

```python
#Removing 'Unrated' row
appDF7 = appDF6.copy()
appDF7 = appDF7.drop(index = appDF7[appDF7.Content_Rating == 'Unrated'].index)

#Combining 'Mature 17+' and 'Adults only 18+' to 'Mature 17+'
appDF7.Content_Rating = appDF7.Content_Rating.replace('Adults only 18+', 'Mature 17+')
appDF7.Content_Rating.value_counts()

#One-Hot encoding of Content_Rating column
appDF7.Content_Rating = appDF7.Content_Rating.astype('category')
appDF7 = pd.get_dummies(appDF7, columns=['Content_Rating'], prefix=['ContRat'])
appDF7.head()
```

```python
appDF7.dtypes
```

```python
#Replacing 'VWD' with mode size, ie., 14000000.0
appDF8 = appDF7.copy()
appDF8.Size = appDF8.Size.replace('VWD', 14000000.0)
appDF8.Size = appDF8.Size.astype(int)
appDF8.dtypes
```

```python
appDF8.head()
```

```python
appDF8.Category.value_counts()
```

```python
#Removing category column
appDF9 = appDF8.iloc[:, 1:]
appDF9.head()
```

```python
appDF9.Rating.unique()
```

```python
installsDict
```

```python
def labelise(df):
    def createLabel(row):
        if(row.Rating >= 4.0 and row.Installs >= 12): row['Label'] = int(1)
        else: row['Label'] = int(0)
        return row
        
    return df.apply(lambda x: createLabel(x), axis=1)
```

```python
#Every column converted to float
labelDF1 = labelise(appDF9)
labelDF1.Label = labelDF1.Label.astype(int)
labelDF1.head(20)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def createXy(df):
    return df.iloc[:, :-1], df.iloc[:, -1:]
```

```python
#No 'Category' column
X1, y1 = createXy(labelDF1)
train_X1, test_X1, train_y1, test_y1 = train_test_split(X1, y1, test_size=0.25)
```

```python
test_X1.head()
```

```python
model1 = LogisticRegression().fit(train_X1, train_y1.values.ravel())
predict_y1 = model1.predict(test_X1)
```

```python
print("%.2f" % model1.score(test_X1, test_y1))
```

```python
print(classification_report(test_y1, predict_y1))
```

```python
print(confusion_matrix(test_y1, predict_y1))
```

```python
#Scaling non 'category' data
labelDF2 = labelDF1.copy()
#Scaling 'Size' column
labelDF2.Size = labelDF2.Size.map(lambda x: x/10000000)
labelDF2.head()
```
